```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
df.sample(10)
```

```python
df = df.iloc[: , 2:]
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(df[['review' , 'education']] , df['purchased'] , test_size = 0.1 , random_state = 0 )
```

```python
x_train
```

```python
from sklearn.preprocessing import OrdinalEncoder 
oe = OrdinalEncoder(categories = [['Poor','Average','Good'] , ['School','UG' , 'PG']])
```

```python
oe.fit(x_train)
x_train_encoded = oe.transform(x_train).astype('int64')
x_test_encoded = oe.transform(x_test).astype('int64')
```

```python
x_train_encoded = pd.DataFrame(x_train_encoded , columns = x_train.columns)
x_test_encoded = pd.DataFrame(x_test_encoded , columns = x_test.columns)
x_train_encoded
```
