# Univariate EDA directly

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python

data = pd.read_csv('train.csv')
data.sample(5)
```

```python
data.describe()
```

## Categorical


```python
sns.countplot(x = 'Survived' ,data = data)
plt.title('Frequency of Survived and Not Survived')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x = 'Sex' , data = data , color = 'orange')
plt.show()
```

```python
valueCount = data['Survived'].value_counts()
valueCount
```

```python
plt.pie(valueCount.values , labels = valueCount.index , autopct = '%1.01f%%')
plt.show()
```

```python
typeofclass = data['Pclass'].value_counts()
type(typeofclass)
plt.pie(typeofclass.values , labels = typeofclass.index , autopct = '%1.02f%%')
```

## Numerical 

```python
data.sample(5)
```

```python
sns.histplot(data = data  , x = 'Age' , bins = 30 , kde = True)
```

```python
sns.displot(data = data , x = 'Age' , bins = 30 , kde = True,height=5, aspect=1.5 )
```

```python
# Probability density function 
sns.displot(x = 'Age', data = data , kde = True ,bins = 10 ,  aspect = 1.5, stat = 'density')
```

```python
sns.boxplot(data = data , x = 'Age')
```

```python
sns.boxplot(data = data , x = 'Fare')
```

```python
sns.boxplot(x = 'Survived' , y = 'Age' , data = data)
plt.title('Box Plot of Age Grouped by Survived')
```
