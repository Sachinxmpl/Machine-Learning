```python
import numpy as np 
from sklearn.datasets import make_classification 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
x , y = make_classification(n_samples = 200 , n_features = 2 , n_informative = 1 ,n_redundant = 0 ,  n_classes = 2 , n_clusters_per_class = 1 , random_state = 41 , hypercube = False , class_sep = 10)
```

```python
def step(x): 
    if x >= 0 : 
        return 1 
    else : 
        return 0 
```

```python
np.insert(x , 0 , 1 , axis = 1)
```

```python
def preceptronTrick(x , y , epoch = 1000 , learning_rate = 0.01) : 
    x = np.insert(x ,0 , 1 , axis =1 ) 
    coefficients = np.ones(x.shape[1])
    
    for i in range(epoch) : 
        random_idx = np.random.randint(0 , x.shape[0])
        y_hat = step(np.dot(coefficients , x[random_idx]))
        coefficients = coefficients + learning_rate * (y[random_idx] - y_hat) * x[random_idx]

    return coefficients[0] , coefficients[1:]





```

```python
bias , weights =preceptronTrick(x , y)
```

```python
print(bias)
print(weights)
```

```python
intercept = - bias / weights[1]
slope = - weights[0]/weights[1]
print(intercept , slope)
```

```python
x_test = np.linspace(-3 ,3 , 100) 
y_pred = slope * x_test + intercept
```

```python

from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
lor.fit(x,y)
```

```python
m = -(lor.coef_[0][0]/lor.coef_[0][1])
b = -(lor.intercept_/lor.coef_[0][1])
```

```python
y_pred_scikitlearn = m * x_test + b
```

```python
plt.figure(figsize = (12,6))
plt.scatter(x[:, 0] , x[: , 1] , c = y, cmap = 'winter' , s = 100)
plt.plot(x_test , y_pred)
plt.plot(x_test , y_pred_scikitlearn)
plt.ylim(-2.5 , 2.6)
```
