```python
sns.boxplot(x = 'Sex'  , y = 'Age' , data = titanic , palette='viridis' , hue = 'Survived')
```

```python
sns.displot(x = 'Age' , data =  titanic , kde = True , stat = 'density' , bins = 10 , aspect=1.5)
```

```python
sns.displot(data = titanic , x = titanic[titanic['Survived'] == 0]['Age'] , kind = 'kde'   , aspect = 1.5 )
sns.displot(data = titanic ,x = titanic[titanic['Survived'] == 1]['Age'] , kind = 'kde' , aspect=1.5)
plt.show()
```

```python
sns.displot(data = titanic , x = 'Age' , hue = 'Survived' , kind = 'kde' ,aspect = 1.5)
```

```python
# Plot both KDEs on the same graph
sns.kdeplot(
    data=titanic[titanic['Survived'] == 0]['Age'],
    label='Survived = 0',
    fill=False,  # Optional: Set to True to fill under the curve
    bw_adjust=1  # Adjust smoothness if needed
)

sns.kdeplot(
    data=titanic[titanic['Survived'] == 1]['Age'],
    label='Survived = 1',
    fill=False,
    bw_adjust=1
)
plt.legend()
plt.title("Probability Density Function")
```
