```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
iris  = sns.load_dataset('iris')
iris
```

```python
sns.kdeplot(data = iris ,x = 'sepal_length')
```

```python
temp = sorted(iris['sepal_length'].tolist())
```

```python
y_quant = []
for i in range(1 , 101):
    y_quant.append(np.percentile(temp , i))
```

```python
sample_normal = np.random.normal(loc = 0 , scale = 1 , size = 1000)
```

```python
x_quant = []
for i in range(1 , 101):
    x_quant.append(np.percentile(sample_normal , i))
```

```python
sns.scatterplot(x = x_quant , y = y_quant)
```
