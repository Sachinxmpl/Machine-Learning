```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('../../datasets/package.csv')
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(data = df , x = 'cgpa' , y = 'package')
plt.title("Scatter plt ")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(df[['cgpa']] , df[['package']] , random_state = 42, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train , y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error

y_pred = lr.predict(x_test)
y_pred = lr.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse * 100}")

```
